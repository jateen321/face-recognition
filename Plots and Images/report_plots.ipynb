```python
def plot_accuracy_bar_graph(data, title):
    features = data['Features']
    train_baseline = data['Baseline_Train']
    test_baseline = data['Baseline_Test']
    train_oversampling = data['Oversampling_Train']
    test_oversampling = data['Oversampling_Test']
    train_augmentation = data['Augmentation_Train']
    test_augmentation = data['Augmentation_Test']
    
    x = np.arange(len(features))
    width = 0.2
    
    plt.figure(figsize=(14, 8))
    plt.bar(x - width, train_baseline, width=width, align='center', label='Baseline (Train)', color='r', alpha=0.5)
    plt.bar(x - width, test_baseline, width=width, align='center', label='Baseline (Test)', color='r', alpha=0.7)
    plt.bar(x, train_oversampling, width=width, align='center', label='Oversampling (Train)', color='g', alpha=0.5)
    plt.bar(x, test_oversampling, width=width, align='center', label='Oversampling (Test)', color='g', alpha=0.7)
    plt.bar(x + width, train_augmentation, width=width, align='center', label='Augmentation (Train)', color='b', alpha=0.5)
    plt.bar(x + width, test_augmentation, width=width, align='center', label='Augmentation (Test)', color='b', alpha=0.7)
    
    plt.xlabel('Features')
    plt.ylabel('Accuracy')
    plt.title(title)
    plt.xticks(x, features, rotation=45, ha='right')
    plt.legend(loc='upper left')
    plt.ylim(0, 135)
    plt.yticks(np.arange(0, 101, 10))
    plt.tight_layout()
    plt.show()
```

```python
# Plotting for the first table
plot_accuracy_bar_graph(df1, "Accuracies on Datasets with min 20 faces per person with 62 persons")

# Plotting for the second table
plot_accuracy_bar_graph(df2, "Accuracies on Datasets with min 100 faces per person with 5 different persons")
```
